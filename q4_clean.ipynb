```python
dec = pd.read_csv("Resources/202312-citibike-tripdata.csv", low_memory=False)
```

```python
q4 = [octo, nov, dec]
```

```python
all = pd.concat(q4, ignore_index=True)
```

```python
all.dropna(subset=['start_station_id', 'start_lat', 'start_lng', 'end_station_id', 'end_lat', 'end_lng'])
```

```python
# Group by start_station_id and calculate mode for start_lat and start_lng
start_df = all.groupby('start_station_id')[['start_lat', 'start_lng']].agg(lambda x: x.mode().iat[0]).reset_index()

# Group by end_station_id and calculate mode for end_lat and end_lng
end_df = all.groupby('end_station_id')[['end_lat', 'end_lng']].agg(lambda x: x.mode().iat[0]).reset_index()
```

```python
all.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng'])
```

```python
# Merge start_df and end_df with the original DataFrame
entire_df = pd.merge(all, start_df, on='start_station_id')
entire_df = pd.merge(entire_df, end_df, on='end_station_id')

# Display the entire DataFrame
entire_df.head()
```

```python
new = entire_df.drop(columns=['start_lat_x','start_lng_x','end_lat_x','end_lng_x'])
```

```python
clean = new.rename(columns={'start_lat_y':'start_lat', 'start_lng_y':'start_lng', 'end_lat_y':'end_lat', 'end_lng_y':'end_lng'})
```

```python
clean.head()
```

```python
max_lat = clean[['start_lat', 'end_lat']].max()
max_lng = clean[['start_lng', 'end_lng']].max()
min_lat = clean[['start_lat', 'end_lat']].min()
min_lng = clean[['start_lng', 'end_lng']].min()
```
