```python
import csv
import pandas as pd
```

```python
df = pd.read_csv("Resources/6_mo_ny_weather.csv")
```

```python
df.head()
print(len(df))
```

```python
df.dropna(inplace=True)
```

```python
print(len(df))
```

```python
df.head()
```

```python
# Assuming you have a DataFrame called df
# Define the maximum and minimum latitude and longitude values
max_lat = 40.8863
min_lat = 40.633385
max_lng = -73.84978
min_lng = -74.086701

# Filter the DataFrame based on the latitude and longitude conditions
filtered_df = df[(df['LATITUDE'] <= max_lat) & (df['LATITUDE'] >= min_lat) & (df['LONGITUDE'] <= max_lng) & (df['LONGITUDE'] >= min_lng)]

filtered_df.head()
```

```python
filtered_df.info()
```

```python

filtered_df['DATE'] = pd.to_datetime(filtered_df['DATE'])

# Filter the DataFrame based on the date range
q4 = filtered_df[(filtered_df['DATE'].dt.month >= 10) & (filtered_df['DATE'].dt.day >= 1) & (filtered_df['DATE'].dt.month <= 12) & (filtered_df['DATE'].dt.day <= 31)]

# Print the filtered DataFrame
q4.head()
```

```python
q4.drop(columns=['ELEVATION', 'STATION'])
```

```python
q4.to_csv("Resources/q4_weather.csv")
```
